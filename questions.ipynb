7. Quels sont les 5 hôtes les plus populaires (ceux dont les propriétés ont reçu le plus de commentaires) ?

```python
question7 = db.airbnb.aggregate(
    [
        {
            "$unwind": "$reviews"
        },
        {
            "$group": {
                "_id": "$name",
                "nombre_commentaires": {"$sum" :1}
            }
        },
        {
            "$sort": {"nombre_commentaires": -1}
        },
        {
            "$limit": 5
        }
    ]
)

pd.DataFrame(list(question7))
```

```python
#Les 5 hôtes les plus populaires sont: "#Private Studio - Waikiki Dream", "Near Airport private room, 2 bedroom", "La Sagrada Familia (and metro) 4 blocks!", "PRIVATE Room in Spacious, Quiet Apt" et "traditional and Charming room".
```

8. Quelles sont les 6 villes ayant le plus de logements disponibles à la location ?

9. Combien de propriétés acceptent plus de 4 invités et ont une caution de moins de 300€ ?

```python
question9 = db.airbnb.aggregate(
    [
        {
            "$match":{
                "accommodates": {"$gt": 4},
                "security_deposit": {"$lt": 300}
            }
        },
        {
            "$count": "nombre_proprietes"
        }
    ]
)

pd.DataFrame(list(question9))
```

```python
#Il y a 451 propriétés qui acceptent plus de 4 invités et ont une caution de moins de 300€.
```

10. Donner les 20 utilisateurs qui ont fait le plus de commentaires (afficher seulement l'ID et le nom de l'utilisateur).

```python
question10 = db.airbnb.aggregate(
    [
        {
            "$unwind": "$reviews"
        },
        {
            "$group": {
                "_id": {
                    "reviewer_id": "$reviews.reviewer_id",
                    "reviewer_name": "$reviews.reviewer_name"
                },
                "nombre_commentaires": {"$sum": 1}
            }
        },
        {
            "$sort": {"nombre_commentaires": -1}
        },
        {
            "$limit": 20
        },
        {
            "$project": {  
                "_id": 0,
                "reviewer_id": "$_id.reviewer_id",
                "reviewer_name": "$_id.reviewer_name"
            }
        }
    ]
)

pd.DataFrame(list(question10))
```

```python
#Les 3 utilisateurs qui ont fait le plus de commentaires sont Filipe, Nick et Uge.
```

11. Parmi les logements à Sydney, quel est la note moyenne des visiteurs ?

```python
question11 = db.airbnb.aggregate(
    [
        {
            "$match": {
                "address.market": "Sydney"
            }
        },
        {
            "$group": {
                "_id": None,
                "note_moyenne_sydney": {"$avg": "$review_scores.review_scores_rating"}
            }
        },
        {
            "$project": {  
                "_id": 0,
                "note_moyenne_sydney" : 1
            }
        }
    ]
)

pd.DataFrame(list(question11))
```

```python
#La note moyenne des visiteurs parmis les logements à Sydney est d'environ 93,5.
```
